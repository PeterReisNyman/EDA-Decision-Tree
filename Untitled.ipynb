```python
def test_model(model,rang):
    scores = []
    for _ in range(rang):
        X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
        model.fit(X_train, Y_train)
        predictions = model.predict(X_test)
        score = accuracy_score(Y_test, predictions)
        scores.append(accuracy_score(Y_test, predictions))

    return scores

cross_val = test_model(model,1000)
print('10 rounds:')
print(f'min: {min(cross_val):.4f} max: {max(cross_val):.4f} avg: {np.mean(cross_val):.4f}')
```

What standardization does:

Standardization (also called Z-score normalization) transforms features so that they have a mean of 0 and a standard deviation of 1. For each feature, it subtracts the mean and divides by the standard deviation: z = (x - ) / 

Why it's useful:
Standardization puts all features on the same scale, preventing features with larger ranges from dominating the model. It helps algorithms that are sensitive to feature scales (like gradient descent-based methods). It makes features comparable and easier to interpret. For decision trees specifically, standardization typically doesn't affect performance since trees make splits based on thresholds, but it can help with visualization and interpretation.

What normalization does:

Normalization scales each sample (row) to have unit norm (length of 1). It divides each sample by its L2 norm (Euclidean length): x_normalized = x / ||x||

How it differs from standardization:
Normalization operates on rows (samples), making each sample vector have length 1, while standardization operates on columns (features), making each feature have mean 0 and std 1. Normalization preserves the direction of the data vector but changes its magnitude. Normalization is useful when the direction of the data matters more than the magnitude (e.g., text classification, recommendation systems). For decision trees, normalization typically has minimal impact since trees split on individual feature values, not on the magnitude of feature vectors.

```python
from imblearn.under_sampling import RandomUnderSampler

def test_rus_model(model,rang):

    rus = RandomUnderSampler(random_state=42)
    X_rus, y_rus = rus.fit_resample(ds['data'], ds['target'])

    scoresTe = []
    scoresTr = []
    for _ in range(rang):
        X_train, X_test, Y_train, Y_test = train_test_split(X_rus, y_rus, test_size=0.2)
        model.fit(X_train, Y_train)
    
        predictions = model.predict(X_train)
        scoresTr.append(accuracy_score(Y_train, predictions))
        predictions = model.predict(X_test)
        scoresTe.append(accuracy_score(Y_test, predictions))
    
    return scoresTe, scoresTr


model = DecisionTreeClassifier(criterion='entropy',max_depth=3, min_samples_split=4, min_samples_leaf=3)
scoresTe, scoresTr = test_rus_model(model,10000)

print(f'--- train ---')
print(f'avg: {np.mean(scoresTr):.4f}')
print(f'\n\n--- test ---')
print(f'avg: {np.mean(scoresTe):.4f}')
```

Undersampling of data means addressing the imbalances of classes in the dataset by removing data from the majority class in the dataset. Should be used when the amount of data is sufficient.

To achieve a general accuracy of approximately 0.93, we combined the following methods:

1. Oversampling (RandomOverSampler): The wine dataset has a slight class imbalance (59, 71, 48 samples for classes 0, 1, 2 respectively). Oversampling creates synthetic copies of minority class samples to balance the dataset, ensuring the model learns all classes equally well.

2. Entropy criterion: Using entropy as the split criterion (instead of Gini) helps the decision tree make more balanced splits and can lead to better generalization.

3. Max depth = 3: Limiting the depth prevents the tree from becoming too complex and overfitting. A shallow tree is more general and interpretable.

4. Min samples split = 4: Requiring at least 4 samples to split a node prevents the tree from creating splits based on very few samples, which could be noise.

5. Min samples leaf = 3: Ensuring each leaf has at least 3 samples prevents overfitting to individual data points and improves generalization.

Why these methods work well together:
Oversampling addresses the class imbalance, giving the model more balanced training data. The hyperparameters (max_depth, min_samples_split, min_samples_leaf) control model complexity, preventing overfitting. Together, they create a model that is general enough to perform well on unseen data (achieving approximately 0.95 accuracy in cross-validation).

What the crosstab shows:

The crosstab (also called confusion matrix) shows the relationship between actual labels (rows) and predicted labels (columns). Each cell (i,j) contains the number of samples with actual label i that were predicted as label j.

General interpretation:
Diagonal elements (top-left to bottom-right) represent correct predictions. Off-diagonal elements represent misclassifications. A perfect model would have all values on the diagonal and zeros elsewhere.

Analysis of this particular case:

Looking at the crosstab where prediction columns are 0, 1, 2 and label rows are 0, 1, 2:
Class 0 had 10 correct predictions, 1 misclassified as class 1, and 0 misclassified as class 2.
Class 1 had 0 misclassified as class 0, 9 correct predictions, and 0 misclassified as class 2.

Class 1 performance: All 9 samples correctly classified (100% accuracy). Perfect classification for this class.

Class 2 performance: 8 out of 9 samples correctly classified (88.9% accuracy). 1 sample misclassified as class 1. No confusion with class 0.

```python
# 10.6. Describe how the numbers (for precision, recall and f1 and general accuracy) in the 
# classification report can be explained by using the cross tab
```

How classification report metrics are calculated from the crosstab:

Using the crosstab where prediction columns are 0, 1, 2 and label rows are 0, 1, 2:
Row 0: 10, 1, 0
Row 1: 0, 9, 0
Row 2: 0, 1, 8

Precision = True Positives / (True Positives + False Positives)
Precision measures: "Of all samples predicted as class X, how many were actually class X?"
Class 0: 10 / (10 + 0) = 10/10 = 1.00 (column 0: all predictions were correct)
Class 1: 9 / (9 + 1 + 1) = 9/11 = 0.82 (column 1: 9 correct, 2 false positives)
Class 2: 8 / (8 + 0) = 8/8 = 1.00 (column 2: all predictions were correct)

Recall = True Positives / (True Positives + False Negatives)
Recall measures: "Of all actual class X samples, how many did we correctly identify?"
Class 0: 10 / (10 + 1 + 0) = 10/11 = 0.91 (row 0: 10 correct, 1 missed)
Class 1: 9 / (0 + 9 + 0) = 9/9 = 1.00 (row 1: all found correctly)
Class 2: 8 / (0 + 1 + 8) = 8/9 = 0.89 (row 2: 8 correct, 1 missed)
