```python
import sklearn
import random
import torch 
import pandas as pd
import  numpy as np
import matplotlib as plt
import joblib as jb
import pydotplus
from matplotlib.pyplot import plot
from IPython.display import Image
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from pandas.plotting import scatter_matrix
from sklearn.tree import DecisionTreeClassifier

ds = sklearn.datasets.load_wine(return_X_y=False, as_frame=False)

ix = [i for i,d in enumerate(ds['feature_names']) if d =='od280/od315_of_diluted_wines']
ds['feature_names'][ix[0]] = 'ratio_of_diluted_wines'

X = ds['data']
Y = ds['target']

plt.rcParams.update({
    "xtick.labelsize": 6,
    "ytick.labelsize": 6
})
```

```python
# Start by building a decision tree for the dataset as it is
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
```

```python
# Use cross validation and describe the performance, in terms of lowest, highest and average performance.
```

```python
predictions = model.predict(X_test)
predictions
```

```python
accuracy_score(Y_test, predictions)
```

```python
def test_model(model,rang):
    scores = []
    for _ in range(rang):
        X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
        model.fit(X_train, Y_train)
        predictions = model.predict(X_test)
        scores.append(accuracy_score(Y_test, predictions))

    return scores

cross_val = test_model(model,1000)
print('10 rounds:')
print(f'min: {min(cross_val):.4f} max: {max(cross_val):.4f} avg: {np.mean(cross_val):.4f}')

```

```python
# Print the tree and analyze it. Describe your analysis, what is good/bad about the tree? What do you see that could be improved? 
# Does your tree have an accuracy that lies within the accuracy results from the cross validation above?
```

```python
jb.dump(model, 'wine-decision-tree.joblib')
model = jb.load('wine-decision-tree.joblib')

dot_data = sklearn.tree.export_graphviz(model, out_file=None, feature_names=ds['feature_names'],label='all', rounded=True, filled=True)

graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
# Improve the tree by adjusting its depth to a limit that you see fit.
```

```python
result = []

for i in range(1,30):
    model = None
    model = DecisionTreeClassifier(max_depth=i)
    cross_val = test_model(model,300)
    print(f'model depths: {i} {np.mean(cross_val):.4f}')
    result.append(cross_val)
    
res_mean = [np.mean(s) for s in result]
print(f'Best depth: {[i+1 for i,s in enumerate(res_mean) if s == max(res_mean)][0]}')
vis = plot([i+1 for i,_ in enumerate(res_mean)], res_mean)
```
