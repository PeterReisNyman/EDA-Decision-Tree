An unbalanced dataset will disrupt learning. The model will more often predict the majority class, this lowers accuracy for the smaller classes.
```python
import sklearn
import random
import torch 
import torch.nn.functional as F
import torch.nn as nn
import pandas as pd
import  numpy as np
import matplotlib as plt
import joblib as jb
import pydotplus
import imblearn
from matplotlib.pyplot import plot
from IPython.display import Image
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from pandas.plotting import scatter_matrix
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier

ds = sklearn.datasets.load_wine(return_X_y=False, as_frame=False)

ix = [i for i,d in enumerate(ds['feature_names']) if d =='od280/od315_of_diluted_wines']
ds['feature_names'][ix[0]] = 'ratio_of_diluted_wines'

X = ds['data']
Y = ds['target']

plt.rcParams.update({
    "xtick.labelsize": 6,
    "ytick.labelsize": 6
})
```

```python
def test_model(model,rang):
    scores = []
    for _ in range(rang):
        X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
        model.fit(X_train, Y_train)
        predictions = model.predict(X_test)
        score = accuracy_score(Y_test, predictions)
        scores.append(accuracy_score(Y_test, predictions))

    return scores

cross_val = test_model(model,1000)
print('10 rounds:')
print(f'min: {min(cross_val):.4f} max: {max(cross_val):.4f} avg: {np.mean(cross_val):.4f}')
```

```python
result = []

for i in range(1,30):
    model = None
    model = DecisionTreeClassifier(max_depth=i)
    cross_val = test_model(model,1000)
    print(f'model depths: {i} {np.mean(cross_val):.4f}')
    result.append(cross_val)
    
res_mean = [np.mean(s) for s in result]
print(f'Best depth: {[i+1 for i,s in enumerate(res_mean) if s == max(res_mean)][0]}')
vis = plot([i+1 for i,_ in enumerate(res_mean)], res_mean)
```

```python
result = []
r = 10
c = 10
min_ = 2
table = np.zeros((r-min_, c-min_))


max_ = 0
mt = []

for y in range(min_,c):
    for x in range(min_,r):
        model = None
        model = DecisionTreeClassifier(criterion='entropy',min_samples_split=x, min_samples_leaf=y)
        cross_val = np.mean(test_model(model,10000))
        print(f'model: (min_samples_split:{x},min_samples_leaf:{y}) {cross_val:.4f}')
        table[x-min_,y-min_] = cross_val

        if cross_val > max_:
            mt = [x,y]
            max_ = cross_val
            print(f'the best: (min_samples_split: {x}, min_samples_leaf: {y}: {cross_val}')
    

print(f'the best: (min_samples_split: {mt[0]}, min_samples_leaf: {mt[1]}')

vis = plt.pyplot.imshow(table, cmap="gray")
```

```python
def test_standardized_model(model,rang):
    scoresTe = []
    scoresTr = []
    standardized_X = preprocessing.scale(ds['data'])
    for _ in range(rang):
        X_train, X_test, Y_train, Y_test = train_test_split(standardized_X, Y, test_size=0.2)
        model.fit(X_train, Y_train)
    
        predictions = model.predict(X_train)
        scoresTr.append(accuracy_score(Y_train, predictions))
        predictions = model.predict(X_test)
        scoresTe.append(accuracy_score(Y_test, predictions))

    return scoresTe, scoresTr

model = DecisionTreeClassifier(criterion='entropy',max_depth=3, min_samples_split=3, min_samples_leaf=3)
scoresTe, scoresTr = test_standardized_model(model,10000)


print(f'--- train ---')
print(f'avg: {np.mean(scoresTr):.4f}')
print(f'\n\n--- test ---')
print(f'avg: {np.mean(scoresTe):.4f}')
print()

colors = {'class_0': 'purple',
          'class_1': 'lightblue',
          'class_2': 'gold'}

c = [colors[f'class_{y}'] for y in Y]
df_standardized = pd.DataFrame(preprocessing.scale(ds['data']), columns=ds.feature_names)
visual = scatter_matrix(df_standardized, figsize=(10,10), c=c, s=20, diagonal='hist')
```

```python
from imblearn.over_sampling import RandomOverSampler

def test_ros_model(model,rang):
    ros = RandomOverSampler(random_state=42)
    X_ros, y_ros = ros.fit_resample(ds['data'], ds['target'])
    
    scoresTe = []
    scoresTr = []
    for _ in range(rang):
        X_train, X_test, Y_train, Y_test = train_test_split(X_ros, y_ros, test_size=0.2)
        model.fit(X_train, Y_train)
    
        predictions = model.predict(X_train)
        scoresTr.append(accuracy_score(Y_train, predictions))
        predictions = model.predict(X_test)
        scoresTe.append(accuracy_score(Y_test, predictions))

    return scoresTe, scoresTr
    
model = DecisionTreeClassifier(criterion='entropy',max_depth=3, min_samples_split=4, min_samples_leaf=3)
scoresTe, scoresTr = test_ros_model(model, 10000)

print(f'--- train ---')
print(f'avg: {np.mean(scoresTr):.4f}')
print(f'\n\n--- test ---')
print(f'avg: {np.mean(scoresTe):.4f}')

# model = DecisionTreeClassifier(criterion='entropy',max_depth=3, min_samples_split=4, min_samples_leaf=3) avg: 0.9526
# model = DecisionTreeClassifier(criterion='entropy', min_samples_split=4, min_samples_leaf=3) avg: 0.9535

```

```python
from imblearn.under_sampling import RandomUnderSampler

def test_rus_model(model,rang):

    rus = RandomUnderSampler(random_state=42)
    X_rus, y_rus = rus.fit_resample(ds['data'], ds['target'])

    scoresTe = []
    scoresTr = []
    for _ in range(rang):
        X_train, X_test, Y_train, Y_test = train_test_split(X_rus, y_rus, test_size=0.2)
        model.fit(X_train, Y_train)
    
        predictions = model.predict(X_train)
        scoresTr.append(accuracy_score(Y_train, predictions))
        predictions = model.predict(X_test)
        scoresTe.append(accuracy_score(Y_test, predictions))
    
    return scoresTe, scoresTr


model = DecisionTreeClassifier(criterion='entropy',max_depth=3, min_samples_split=4, min_samples_leaf=3)
scoresTe, scoresTr = test_rus_model(model,10000)

print(f'--- train ---')
print(f'avg: {np.mean(scoresTr):.4f}')
print(f'\n\n--- test ---')
print(f'avg: {np.mean(scoresTe):.4f}')
```
